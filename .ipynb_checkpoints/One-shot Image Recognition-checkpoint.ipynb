## Preprocessing - Scale and Resize

```python
def preprocess(file_path):
    image_byte = tf.io.read_file(file_path)
    # decode the mage
    image = tf.io.decode_jpeg(image_byte)
    # resize and rescale
    image = tf.image.resize(image,(105,105))
    image = image / 255.0
    return image
    
anchor_ds = negative.map(preprocess)
```

```python
plt.imshow(next(iter(anchor_ds)).numpy())
```

## Create Labelled Dataset

```python
positives = tf.data.Dataset.zip((anchor,positive, tf.data.Dataset.from_tensor_slices(tf.ones(len(anchor)))))
negatives = tf.data.Dataset.zip((anchor,negative, tf.data.Dataset.from_tensor_slices(tf.zeros(len(anchor)))))
data = positives.concatenate(negatives)
```

```python
next(negatives.as_numpy_iterator())
```

## Build Train and Test Partition

```python
# Preprocessing function for all 3
def preprocess_twin(input_img, validation_img, label):
    return (preprocess(input_img), preprocess(validation_img), label)
```

```python
# pipeline for data loading
data = data.map(preprocess_twin).cache().shuffle(buffer_size=1024)
```

```python
# splitting into train and test data
train_data = data.take(round(len(data)*0.7)).batch(16).prefetch(8)
```

```python
# Testing Partition
test_data = data.skip(round(len(data)*0.7))
test_data = test_data.take(round(len(data)*0.3)).batch(16).prefetch(8)
```

## Model Engineering

## Building Embedding Layer

```python
inp = Input(shape=(105,105,3),name="input_image")
# block 1
conv_1 = Conv2D(64,(10,10),activation="relu")(inp)
max_1 = MaxPooling2D(64,(2,2),padding="same")(conv_1)
# block 2
conv_2 = Conv2D(128,(7,7),activation="relu")(max_1)
max_2 = MaxPooling2D(64,(2,2),padding="same")(conv_2)
# block 3
conv_3 = Conv2D(128,(4,4),activation="relu")(max_2)
max_3 = MaxPooling2D(64,(2,2),padding="same")(conv_3)
# block 4
conv_4 = Conv2D(256,(4,4),activation="relu")(max_3)
# Flatten 
flat = Flatten()(conv_4)
dense_1 = Dense(4096,activation="sigmoid")(flat)
```

```python
def make_embedding():
    inp = Input(shape=(105,105,3),name="input_image")
    # block 1
    conv_1 = Conv2D(64,(10,10),activation="relu")(inp)
    max_1 = MaxPooling2D(64,(2,2),padding="same")(conv_1)
    # block 2
    conv_2 = Conv2D(128,(7,7),activation="relu")(max_1)
    max_2 = MaxPooling2D(64,(2,2),padding="same")(conv_2)
    # block 3
    conv_3 = Conv2D(128,(4,4),activation="relu")(max_2)
    max_3 = MaxPooling2D(64,(2,2),padding="same")(conv_3)
    # block 4
    conv_4 = Conv2D(256,(4,4),activation="relu")(max_3)
    # Flatten 
    flat = Flatten()(conv_4)
    dense_1 = Dense(4096,activation="sigmoid")(flat)
    
    # dense_2 = Dense(1,activation="sigmoid")(flat)
    return Model(inputs=[inp],outputs=[dense_1],name="embedding")
```

```python
embedding = make_embedding()
```

## Build Distance layer

```python
class L1Dist(Layer):
    def __init__(self,**kwargs):
        super(L1Dist,self).__init__()

    def call(self,input_embedding,validation_embedding):
        return tf.math.abs(input_embedding - validation_embedding)
```

```python
l1 = L1Dist()
```

```python
embedding(Input(shape=(105,105,3),name="input_image"))
```

## Make Siamese model

```python
def make_saimese_model():
    input_image = Input(shape=(105,105,3),name="input_image")
    validation_image = Input(shape=(105,105,3),name="validation_image")

    siamese_layer = L1Dist()
    siamese_layer._name = "distance"
    distance = siamese_layer(embedding(input_image),embedding(validation_image))

    output = Dense(1,activation="sigmoid")(distance)
    return Model(inputs=[input_image,validation_image],outputs=[output],name = "Siamese_Network")
```

```python
siamese_model = make_saimese_model()
```

```python
siamese_model.summary()
```

# Training

## Setup Loss and optimizer

```python
bce = tf.keras.losses.BinaryCrossentropy()
```

```python
optimizer = tf.keras.optimizers.Adam(1e-4)
```

## Establish Checkpoints

```python
checkpoint_dir = os.path.join("training_checkpoints","ckpt")
if not os.path.exists(checkpoint_dir):
    os.makedirs(checkpoint_dir)
checkpoint = tf.train.Checkpoint(optimizer=optimizer,siamese_model=siamese_model)
# save_path = checkpoint.save(checkpoint_dir)
```

## Build Train Step Functions

```python
@tf.function
def train_step(batch):
    with tf.GradientTape() as tape:
        # get anchor and positive/negative images
        X = batch[:2]
        # get lavel
        y = batch[2]
        #Forward pass
        y_pred = siamese_model(X,training=True)
        loss_value = bce(y, y_pred)
        
    # print(loss_value)
    grad = tape.gradient(loss_value,siamese_model.trainable_weights)
    optimizer.apply_gradients(zip(grad,siamese_model.trainable_weights))
    # return loss
    return loss_value
```

## Build trainig Loops

```python
def train(train_data,EPOCHS):
    # Loop through epochs
    for epoch in range(EPOCHS):
        print("\n Epoch: {}/{}".format(epoch+1,EPOCHS))
        progbar = tf.keras.utils.Progbar(len(train_data))
        # loop through each button
        for step, batch_train in enumerate(train_data):
            train_step(batch_train)
            progbar.update(step + 1)
        # Save Checkpoints
        if epoch % 10 == 0:
            checkpoint.save(checkpoint_dir)
```

```python
def verify(model,detection_threshold,verification_threshold):
    results = []
    for ver_img_path in verification_dir.glob("*"):
        ver_image = preprocess(os.path.abspath(ver_img_path))
        input_image = preprocess(os.path.join("application_data","input_images","input_img.jpg"))
        inputs = [tf.expand_dims(input_image,axis=0),tf.expand_dims(ver_image,axis=0)]
        result = model.predict(inputs)
        results.append(result)
    num_results_thershold = np.sum(np.array(results) > detection_threshold)
    verified_ratio = num_results_thershold / len(list(verification_dir.glob("*")))
    verified = verified_ratio > verification_threshold
    return results, verified
```
